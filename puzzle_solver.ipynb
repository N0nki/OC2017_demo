```python
import networkx as nx
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspace
from graphillion import GraphSet

from lib.gridgraph import GridGraph

%matplotlib inline

mpl.rcParams['font.family'] = 'Source Han Code JP'
```

```python
# 既に絵が表示されている箇所を格子グラフのノード番号とする
QUESTIONS = {
    "sample": [4, 25]
}
```

```python
m, n = 6, 6 # パズルのサイズ
s, t = 1, (m+1)*(n+1) # スタート，ゴール
```

```python
def solve(target, paths, question):
    for node in question:
        paths = paths.excluding(node)
    paths = paths.graph_size(target - len(question) - 1)
    # print(target - len(question) -1)
    return paths
```

```python
G = GridGraph(m, n)
GraphSet.set_universe(G.nx_graph.edges(), traversal='as-is')
plt.figure(figsize=(m, n))
plt.title("{}×{}の格子グラフ\n頂点の数: {} 辺の数: {}".format(m, n, len(G.nx_graph.nodes()), len(G.nx_graph.edges())))
G.draw()
```

```python
paths = GraphSet.paths(s, t)
answers = solve(t, paths, QUESTIONS["sample"])
len(answers)
```

```python
G.draw(subgraph=answers.choice())
```
